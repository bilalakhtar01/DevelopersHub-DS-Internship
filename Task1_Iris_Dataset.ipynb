```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

print("✅ All libraries imported successfully!")
```

```python
# Step 2: Iris Dataset load kar rahe hain
# Seaborn ke paas yeh dataset already hota hai, download ki zaroorat nahi
df = sns.load_dataset('iris')

print("Dataset load ho gaya! ✅")
print("Dataset ki pehli 5 rows:")
df.head()
```

```python
# Step 3: Dataset ka structure samajhte hain

# Kitni rows aur columns hain?
print("Shape (rows, columns):", df.shape)
print()

# Column ke naam kya hain?
print("Column Names:", df.columns.tolist())
print()

# Har column ka data type aur kitni values hain
print("Dataset Info:")
print(df.info())
print()

# Numbers ka summary (average, min, max waghaira)
print("Statistical Summary:")
df.describe()
```

```python
# Step 4: Scatter Plot - Do variables ka relationship dekhte hain
# Sepal Length vs Sepal Width ka graph bana rahe hain

plt.figure(figsize=(10, 6))   # Graph ka size set kar rahe hain

sns.scatterplot(
    data=df, 
    x='sepal_length',       # X-axis pe sepal length
    y='sepal_width',        # Y-axis pe sepal width
    hue='species',          # Har species ka alag colour
    style='species',        # Har species ki alag shape
    s=100                   # Dots ka size
)

plt.title('Sepal Length vs Sepal Width (by Species)', fontsize=16)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.legend(title='Species')
plt.grid(True, alpha=0.3)
plt.show()

print("Scatter plot ban gaya! ✅")
```

```python
# Step 5: Petal Length vs Petal Width ka scatter plot

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df, 
    x='petal_length', 
    y='petal_width', 
    hue='species',
    style='species',
    s=100
)

plt.title('Petal Length vs Petal Width (by Species)', fontsize=16)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.legend(title='Species')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 6: Histogram - Data distribution dekhte hain
# Har column ka histogram ek saath bana rahe hain

fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 4 graphs ek saath

# Histogram 1 - Sepal Length
axes[0, 0].hist(df['sepal_length'], bins=20, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Sepal Length Distribution')
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Count')

# Histogram 2 - Sepal Width
axes[0, 1].hist(df['sepal_width'], bins=20, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Sepal Width Distribution')
axes[0, 1].set_xlabel('Sepal Width (cm)')
axes[0, 1].set_ylabel('Count')

# Histogram 3 - Petal Length
axes[1, 0].hist(df['petal_length'], bins=20, color='salmon', edgecolor='black')
axes[1, 0].set_title('Petal Length Distribution')
axes[1, 0].set_xlabel('Petal Length (cm)')
axes[1, 0].set_ylabel('Count')

# Histogram 4 - Petal Width
axes[1, 1].hist(df['petal_width'], bins=20, color='plum', edgecolor='black')
axes[1, 1].set_title('Petal Width Distribution')
axes[1, 1].set_xlabel('Petal Width (cm)')
axes[1, 1].set_ylabel('Count')

plt.tight_layout()   # Graphs overlap na hon
plt.show()

print("Histograms ban gaye! ✅")
```

```python
# Step 7: Box Plot - Outliers aur spread dekhte hain

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Box Plot 1
sns.boxplot(data=df, x='species', y='sepal_length', ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Sepal Length by Species')

# Box Plot 2
sns.boxplot(data=df, x='species', y='sepal_width', ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Sepal Width by Species')

# Box Plot 3
sns.boxplot(data=df, x='species', y='petal_length', ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Petal Length by Species')

# Box Plot 4
sns.boxplot(data=df, x='species', y='petal_width', ax=axes[1, 1], palette='Set2')
axes[1, 1].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()

print("Box Plots ban gaye! ✅")
```

```python
# Step 8: Pair Plot - Sab variables ka relationship ek saath
sns.pairplot(df, hue='species', palette='Set1', diag_kind='hist')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02, fontsize=16)
plt.show()

print("Pair Plot ban gaya! ✅")
```

```python
# Step 9: Correlation Heatmap
plt.figure(figsize=(8, 6))

# Sirf numeric columns ka correlation nikalte hain
numeric_df = df.select_dtypes(include='number')
correlation = numeric_df.corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, fmt='.2f')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()

print("Heatmap ban gaya! ✅")
```

## Conclusion

### Key Findings from Iris Dataset Analysis:

1. **Dataset Overview**: The Iris dataset contains 150 samples of 3 flower species (Setosa, Versicolor, Virginica) with 4 numerical features each. There are no missing values.

2. **Scatter Plots**: 
   - Setosa is clearly separable from Versicolor and Virginica in all feature combinations.
   - Petal features provide better separation between species than sepal features.

3. **Histograms**: 
   - Setosa has significantly smaller petal length and petal width compared to other species.
   - Sepal width shows the most overlap between species.

4. **Box Plots**: 
   - Some outliers were detected in sepal width for the Setosa species.
   - Virginica has the largest petal measurements overall.

5. **Correlation Analysis**: 
   - Petal length and petal width have a very strong positive correlation (0.96).
   - Sepal width shows weak/negative correlation with petal features.

### Overall Insight:
Petal measurements (length and width) are more useful than sepal measurements for distinguishing between iris species. Setosa is the most distinct species and can be easily separated from the other two.
